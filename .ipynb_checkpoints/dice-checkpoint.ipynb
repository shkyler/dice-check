### Introduction

It is proposed to carry out an experiment in order to determine whether or not a standard six-sided dice is fair. A fair dice is defined as one where there is an
equal chance of each of the 6 numbers being rolled on each roll of the dice. 

### Equipment Required

1. A dice
1. A cup to roll it from
1. A notebook and pen to record results
1. A computer with the ananconda distribution of python installed to record and an analyse the data from the experiment

### Method

1. Put the dice in the cup
1. Shake the cup vigorously
1. Turn out the dice on a table
1. Record the number that was rolled
1. Repeat steps an equal amount of times e.g. 60 times
1. Tally the results

### Results
The results in the tally should look like this:

<img src="img/results.jpg" width=500>

```python
# import the require librabries
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(20,10))

# record the lists of results and numbers
numbers = ['1','2','3','4','5','6']
results = [5,13,9,11,14,8]

# plot the results
xlocs, xlabs = plt.xticks()
xlocs=[i for i in range(0,6)]
y_pos = np.arange(len(results))
plt.bar(y_pos,height=results, align='center')
plt.xticks(y_pos, numbers)
plt.ylabel('Number of Rolls')
plt.title('Count of each Number Rolled')
for i, v in enumerate(results):
    plt.text(int(numbers[i]) - 1, v + 0.1, str(v))
plt.show()
```

```python
plt.figure(figsize=(20,10))
# Data to plot
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red', 'blue']

# Plot
plt.pie(results, labels=numbers, colors=colors,autopct='%1.1f%%', startangle=140)
plt.title('Percentage of each Number Rolled')
plt.axis('equal')
plt.show()
```
